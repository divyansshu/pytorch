# Creating a Tensor

```python
# using empty
a = torch.empty(2,3)
# allocates space in memory of (2,3) shape
a
```

```python
# check type
type(a)
```

```python
# using zeroes
torch.zeros(2,3)
# creates an array of (2,3) shape with value zero
```

```python
# using ones
torch.ones(2,3)
# creates an array of shape (2,3) of value 1
```

```python
# using rand
torch.rand(2,3)
# creates an array of shape(2,3) with random values between 0 and 1
```

```python
# rand creates arrays with different values at each different call
torch.rand(2,3)
```

```python
# use of seed
# seed does not let change the values of arrays using rand at each different call, multiple users can now access same array values using rand keyword, but seed value must be same

torch.manual_seed(100)
print(torch.rand(2,3))

torch.manual_seed(100)
print(torch.rand(2,3))
```

```python
# using tensor
torch.tensor([[1,2,3], [4,5,6]])
# using tensor we can create any kind of array 1d,2d,Nd..
```

```python
# arange
print('using range ->', torch.range(0,10,2))
```

```python
# using linspace
print('using linspace ->', torch.linspace(10,20,10))
# linspace is linearly space, it creates evenly spaced 1d array
```

```python
# using eye
print('using eye ->', torch.eye(3))
# eye is identity matrix, it creates an identity matrix of given size
```

```python
# using full
print('using full ->', torch.full((3,3),5))
# full creates a matrix of 3 x 3 of number 5
```

# Mathematical Operations

## 1. Scalar operation

```python
x = torch.tensor([[1,2,3],[1,2,3]])
x
```

```python
# addition
x + 2
```

```python
# subtraction
x - 2
```

```python
# multiplication
x * 2
```

```python
# division
x / 2
```

```python
# int division
x // 2
```

```python
# mod
x % 2
```

```python
# power
x**2
```

## 2. Element wise operation

```python
a = torch.tensor([[2,4,6], [2,4,6]])
b = torch.tensor([[1,2,3],[1,2,3]])
print(a)
print(b)
```

```python
# addition
a + b
```

```python
# subtraction
a - b
```

```python
# multiply
a * b
```

```python
# division
a / b
```

```python
# power
a ** b
```

```python
# mod
a % b
```

```python
z = torch.tensor([-1,-2,3,4])
z
```

```python
torch.abs(z)
```

```python
torch.neg(z)
```

```python
d = torch.tensor([1.9, 2.3, 3.7, 4.4])
d
```

```python
torch.round(d)
```

```python
torch.ceil(d)
```

```python
torch.floor(d)
```

```python
# clamp
torch.clamp(d, min=2, max=3) # clamp can be used to set range
```

## 3. Reduction operation

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

```python
# sum
torch.sum(e) # calculate sum of all elements of tensor
```

```python
# sum along columns
torch.sum(e, dim=0)
```

```python
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e)
```

```python
# mean along col
torch.mean(e, dim=0)
```

```python
# mean along rows
torch.mean(e, dim=1)
```

```python
# median
torch.median(e)
```

```python
# max
torch.max(e)
```

```python
# min
torch.min(e)
```

```python
# product
torch.prod(e)
```

```python
# std
torch.std(e)
```

```python
# variance
torch.var(e)
```

```python
# argmax
torch.argmax(e) # returns the index of max element in tensor
```

```python
# argmin
torch.argmin(e) # returns the index of min element in tensor
```
