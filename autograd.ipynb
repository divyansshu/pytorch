```python
import torch
import torch.nn as nn
```

# Create Leaf Tensors

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

# Forward Pass

```python
z = w * x  + b # 2.0 * 3.0 + 0.0 = 6.0
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
criterion = nn.BCELoss()
loss = criterion(y_pred, y)
loss
```

# call.backward()

```python
# Calling .backward() on the loss seeds the backward pass with gradient 1.0,  then walks the graph in reverse — applying the chain rule at each grad_fn. After this, .grad is populated only on leaf tensors that had requires_grad=True. Intermediate tensors (z, y) don't store grads by default.
loss.backward() # walk the graph in reverse

# gradients are now populated on leaf tensors
print(w.grad) # tensor(28.)
print(b.grad) # tensor(42.)
```

# Clearing Grad

```python
x = torch.tensor(2.0, requires_grad=True)
x
```

```python
y = x ** 2
y
```

```python
y.backward()
```

```python
x.grad # first time was 4. and second time was 8.
# calling y.backward() multiple times causes the previous grad to add on to the next grad and changes the next cycle gradients,so it is necessary to clear grad before calling backward() next time.
```

```python
x.grad.zero_()
# grad.zero_() clear the previous gradients, this should be called before backward()
```

# Disable Gradient tracking

```python
x = torch.tensor(2.0, requires_grad=True)
x
```

```python
y = x ** 2
y
```

```python
y.backward()
```

```python
x.grad
```

```python
# if we are done training, with backward pass and now we are predicting the result then, we don't need backward pass, then we can pause or stop the gradients calculation.
# it saves memory and speed
```

# Methods to pause the calculation of gradients
* option 1: requires_grad_(False)
* option 2: detach()
* option 3: torch.no_grad()

## most convienent and mostly used method is option 3.

```python
x = torch.tensor(5.0, requires_grad=True)
x
```

```python
# during inference(validation, testing, deployement), yod don't need gradients- you are just running a forward pass. Wrapping it in torch.no_grad() tells pytorch to skip building the computation graph entirely. This saves the significant memory and speeds up the forward pass. Always use it outside training loops.

with torch.no_grad(): # Inference — no graph needed, saves memory & speed
  y = x ** 2

y
```

```python
# now backward() function will not work because of torch.no_grad() function.
y.backward()
```
